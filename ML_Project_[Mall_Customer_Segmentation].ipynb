
This project performs exploratory data analysis on the **Mall Customers Datase**t to identify patterns in customer demographics, spending behavior, and income distribution.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(style="whitegrid")
```

We load the dataset and check for null values, data types, and basic statistics.

```python
df = pd.read_csv("Mall_Customers.csv")

df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.drop('CustomerID', axis=1, inplace=True)
```

The graphs below show:
1. the count of male vs female customers.
2. the age range of the customers and detetct any skewness
3. highlights the income levels of mall customers
4. spending score indicates customer purchase behavior

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for feature in features:
  sns.histplot(df[feature], kde=True, bins=20)
  plt.title(f'Distribution of {feature}')
  plt.show()
```

#Spending Score vs Income
Analyzing how income affects spending behavior

#Age vs Spending Score
Do younger or older customers tend to spend more?

```python
for feature in features:
  sns.boxplot(x=df[feature])
  plt.title(f'Boxplot of {feature}')
  plt.show()
```

```python
sns.boxplot(x='Gender', y='Spending Score (1-100)', data=df) #leave a whitespace between Score and (1-100)
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue ='Gender', data=df)
plt.title('Income vs Spending Score')
plt.show()
```

```python
#Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

#Insights
1. The dataset has an almost even gender split
2. Younger customers (under 40) tend to spend more.
3. Annual income does not directly correlate with spending score, thus suggesting that marketing should target behavioral segments, and not just income brackets.
4. Middle-income customers exhibit the widest range in spending; they are the sweet spot for targeted promotions.
